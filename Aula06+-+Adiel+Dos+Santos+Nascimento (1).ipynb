<font size="+1"> Resolução do sistema linear mostrado anteriormente  </font>

```python
a = np.array([[2,1],[1,-3]])
b = np.array([3,-2])

x = np.linalg.solve(a,b)
print(x)

print(np.allclose(np.dot(a,x),b))
```

<font size="+1"> Reproduzir Gráfico mostrado na aula  </font>

```python
A  = np.matrix([[2, 1], [1, -3]])
#print(a)

vetor1  = np.array([1, 1])
#print('v', v)


alfa = vetor1[0] / vetor1[1]
x = np.array(range(0, vetor1[0]+1))
y = alfa * x
plt.plot(x,y)

vetor2 = np.matmul(A.data,vetor1)
alfa2 = vetor2[0] / vetor2[1]

x = np.array(range(0, vetor2[0]))
y = alfa2 * x
plt.plot(x,y)

plt.show()
```

<font size="+1"> Reproduzir Gráfico com A multiplicado por 2  </font>

```python
A  = np.matrix([[2, 1], [1, -3]])
#print(a)

vetor1  = np.array([1, 1])
#print('v', v)


alfa = vetor1[0] / vetor1[1]
x = 2 * (np.array(range(0, vetor1[0]+1)))
y = alfa * x
plt.plot(x,y)

vetor2 = np.matmul(A.data,vetor1)
alfa2 = vetor2[0] / vetor2[1]

x = np.array(range(0, vetor2[0]))
y = alfa2 * x
plt.plot(x,y)

plt.show()
```
